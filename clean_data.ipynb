```python
import pandas as pd

# đọc file csv
df = pd.read_csv('data_deep_clean_v4.csv')
df
```

```python
# Sắp xếp theo 'DotHoc' nhưng giữ nguyên thứ tự 'IDSinhVien'
df = df.sort_values(by=['IDSinhVien', 'DotHoc'], kind='mergesort')
```

```python
# xóa cột MaHoa_HocKy_NamHoc
df = df.drop(columns=['MaHoa_HocKy_NamHoc'])
```

```python
# Hàm mã hóa xếp loại dựa trên DiemTongKet
def ma_hoa_xep_loai(diem):
    if 9.0 <= diem <= 10.0:
        return 5
    elif 8.5 <= diem <= 8.9:
        return 4
    elif 7.0 <= diem <= 8.4:
        return 3
    elif 5.5 <= diem <= 6.9:
        return 2
    elif 4.0 <= diem <= 5.4:
        return 1
    elif 0.0 <= diem <= 3.9:
        return 0

# Áp dụng hàm để tạo cột mới 'MaHoa_XepLoai'
df['MaHoa_XepLoai'] = df['DiemTongKet'].apply(ma_hoa_xep_loai)
```
