```python
import pandas as pd

# đọc dữ liệu từ file csv
df = pd.read_csv('data_deep_clean_v6.csv')
df
```

# `Biểu đồ phân bố điểm tổng kết`

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=df, x='DiemTongKet', kde=True, bins=10, color = 'red', edgecolor = 'black', linewidth = 1.2)
plt.title('Distribution of Final Grades')
plt.xlabel('Final Grade')
plt.ylabel('Count')


plt.show()
```

# `Biểu đồ điểm theo môn học`

```python
avg_grades = df.groupby('TenMonHoc')['DiemTongKet'].mean().sort_values()
avg_grades.plot(kind='barh', figsize=(10,8), color='skyblue')
plt.title('Average Final Grade by Subject')
plt.xlabel('Average Final Grade')
plt.ylabel('Subject')

# hiện số điểm trung bình trên cột
for index, value in enumerate(avg_grades):
    plt.text(value, index, round(value, 2))

plt.show()
```

# `Biểu đồ điểm theo kỳ học (HK)`

```python
sns.boxplot(data=df, x='DotHoc', y='DiemTongKet')
plt.title('Final Grade Distribution by Semester')
plt.xlabel('Semester')
plt.ylabel('Final Grade')
plt.show()
```

# `Biểu đồ phân bố điểm theo giới tính`

```python
sns.violinplot(data=df, x='GioiTinh', y='DiemTongKet')
plt.title('Final Grade Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Final Grade')
plt.show()
```

# `Biểu đồ điểm theo lớp học`

```python
# check số lượng các giá trị khác nhau 
df['IDLopHoc'].nunique()
```

```python
avg_grades_by_class = df.groupby('IDLopHoc')['DiemTongKet'].mean().sort_values()
avg_grades_by_class.plot(kind='barh', figsize=(10,8), color='lightgreen')
plt.title('Average Final Grade by Class')
plt.xlabel('Average Final Grade')
plt.ylabel('Class ID')


for index, value in enumerate(avg_grades_by_class):
    plt.text(value, index, round(value, 2))
    
plt.show()
```

# `Biểu đồ phân tán điểm theo các môn học cúng giới tính`

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='TenMonHoc', y='DiemTongKet', hue='GioiTinh')
plt.title('Grade Distribution by Subject')
plt.xticks(rotation=90)
plt.xlabel('Subject')
plt.ylabel('Final Grade')
plt.show()
```

# `Biểu đồ phân bố điểm theo từng ID lớp`

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lấy danh sách các lớp học duy nhất
unique_classes = df['IDLopHoc'].unique()


for class_id in unique_classes:
    class_data = df[df['IDLopHoc'] == class_id]
    
    plt.figure(figsize=(10, 6))
    sns.barplot(data=class_data, x='TenMonHoc', y='DiemTongKet', palette='viridis')
    plt.title(f'Final Grade Distribution for Class ID: {class_id}')
    plt.xlabel('Subject Code')
    plt.ylabel('Final Grade')
    plt.xticks(rotation=90)
    plt.show()
```

# `Tỷ lệ qua môn và không qua môn tổng`

```python
status_counts = df['QuaMon'].value_counts()
status_percent = status_counts / status_counts.sum() * 100

plt.figure(figsize=(8, 8))
plt.pie(status_percent, labels=status_percent.index, autopct='%1.1f%%', colors=['red', 'green'], startangle=140)
plt.title('Percentage of Students Passing and Failing')
plt.show()
```

# `Số lượng sinh viên qua môn và ko qua môn theo các môn học`

```python
import pandas as pd
import matplotlib.pyplot as plt

mon_stats = df.groupby(['TenMonHoc', 'QuaMon']).size().unstack(fill_value=0)

ax = mon_stats.plot(kind='bar', stacked=True, figsize=(8, 18), color=['red', 'green'])

plt.title('Number of Students Failing and Passing by Subject')
plt.xlabel('Subject Code (TenMonHoc)')
plt.ylabel('Number of Students')
plt.legend(['Fail (Rớt Môn)', 'Pass (Qua Môn)'], title='Status')
plt.xticks(rotation=90)


for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10, padding=3)


plt.tight_layout()
plt.show()
```
