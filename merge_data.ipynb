```python
import pandas as pd

# để duowgf dẫn đọc file
file_path = 'THONGKESVISHOCTUCHON.xlsx'

sheet_dstk = pd.read_excel(file_path, sheet_name='DSTK')
sheet_ds_mon_tc = pd.read_excel(file_path, sheet_name='DS môn TC', header=1)
sheet_ds_co_diem = pd.read_excel(file_path, sheet_name='DS co Diem')
sheet_dulieu0205 = pd.read_excel(file_path, sheet_name='Dulieu0205')

print("Dữ liệu từ sheet DSTK:")
print(sheet_dstk.head())  

print("\nDữ liệu từ sheet DS môn TC:")
print(sheet_ds_mon_tc.head())

print("\nDữ liệu từ sheet DS co Diem:")
print(sheet_ds_co_diem.head())  

print("\nDữ liệu từ sheet Dulieu0205:")
print(sheet_dulieu0205.head()) 
```

```python
# đổi tên lại
sheet_ds_mon_tc = sheet_ds_mon_tc.rename(columns={'Mã môn': 'MaMonHoc'})

# gọp lại
sheet_dulieu0205 = pd.merge(sheet_dulieu0205, sheet_ds_mon_tc[['MaMonHoc', 'Tên Môn', 'HK']], on='MaMonHoc', how='left')

sheet_dulieu0205.head(15)
```

```python
# thêm vào mấy số 0 cho cột mã môn học
sheet_dulieu0205['MaMonHoc'] = sheet_dulieu0205['MaMonHoc'].astype(str).str.zfill(6)

sheet_dulieu0205.head(15)
```

```python
# kiểm tra null
print(sheet_dulieu0205.isnull().sum())

# xóa các dòng có giá trị null
sheet_dulieu0205 = sheet_dulieu0205.dropna(subset=['DiemTongKet'])

# kiểm tra lại
print(sheet_dulieu0205.isnull().sum())
```

```python
sheet_dulieu0205
```

```python
# Tạo cột 'QuaMon' dựa trên điều kiện 'DiemTongKet'
sheet_dulieu0205['QuaMon'] = sheet_dulieu0205['DiemTongKet'].apply(lambda x: 1 if x >= 3.9 else 0)
```

```python
# Tính trung bình điểm cho từng môn học (MaMonHoc)
average_scores = sheet_dulieu0205.groupby('MaMonHoc')['DiemTongKet'].transform('mean')

# làm tròn đến số thập phân thứ 2
average_scores = average_scores.round(2)

# Thêm cột 'DiemTBMon' vào DataFrame
sheet_dulieu0205['DiemTBMon'] = average_scores

```

```python
# Tách cột 'TenDot' thành 2 cột: 'TenDot' và 'NamHoc'
sheet_dulieu0205[['TenDot', 'NamHoc']] = sheet_dulieu0205['TenDot'].str.split(' ', n=1, expand=True)

# Xóa dấu ngoặc () trong cột 'NamHoc'
sheet_dulieu0205['NamHoc'] = sheet_dulieu0205['NamHoc'].str.replace(r'[\(\)]', '', regex=True)
```

```python
# 1. Đổi tên các cột
sheet_dulieu0205 = sheet_dulieu0205.rename(columns={
    'TenDot': 'HocKy',  # Ví dụ: đổi 'TenDot' thành 'HocKy'
    'Tên Môn': 'TenMonHoc',  # Đổi 'Tên Môn' thành 'TenMonHoc'
    'Thuộc KCNTT': 'ThuocKCNTT',  # Đổi 'Thuộc KCNTT' thành 'IDLopHoc'
    # Bạn có thể thêm các cột khác nếu muốn đổi tên
})

# 2. Di chuyển vị trí các cột
# Giả sử bạn muốn đưa cột 'HocKy' và 'NienKhoa' lên đầu
cols = ['IDLopHoc', 'IDSinhVien', 'MaMonHoc', 'HocKy', 'NamHoc', 'DiemTongKet', 'QuaMon', 'DiemTBMon', 'TenMonHoc', 'ThuocKCNTT', 'HK', 'GioiTinh']
sheet_dulieu0205 = sheet_dulieu0205[cols]
```
